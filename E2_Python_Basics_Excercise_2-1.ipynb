### <span style="color: red"> Task 3:</span> With the help of matminer, 1. Retrive the 'steel_strength' dataset into a dataframe and clean your dataset in needed; 2. Featurize the compositions of the materials; and 3. Perform correlation analysis and drop feature(s) with an coefficient greater than abs(0.8) from your dataframe.

### <span style="color: blue"> Solution:</span>

```python
import matminer
from matminer.datasets import load_dataset


#matminer.datasets.get_available_datasets() #you can check what datasets are available with this line
steel = load_dataset('steel_strength')
print(type(steel),'\n')
steel.head(3)
#df2 = StrToComposition().featurize_dataframe(df2, "formula")
```

```python
## add the comp_normalization function from E1, with small modification

## letter, number, symbol
from pymatgen.core.periodic_table import Element
from matminer.featurizers.conversions import StrToComposition

# obtain the list of elements
elem_lst=[]
for i in range(1,118):
    elem_lst.append(Element.from_Z(i).symbol)
#print(elem_lst)

def comp_normalization(name):
    element = []
    number = []
    e=''
    n=''
    for i in name:
        if i.isalpha()==False:
            n+=i
            if e!='':
                element.append(e)
            e=''
        else:
            e+=i
            if n!='':
                number.append(float(n))
            n=''
    number.append(float(n)) #int -> float
    
    ######################################## element name check for BQ #######################################
    for i in element:
        if (i in elem_lst)==False:
            print('ERROR:invalid element name: ',i)
    ##########################################################################################################
    
    # normalization check
    sum_comp=0
    
    if sum(number)==1:
        return name
    else:
        tot = sum(number)  
#         if abs(tot-1)>1e-3:      # 1e-3 is the tolerance for rounding to 3 significant digit
#             print('The composition is not normalized')
        for i,n in enumerate(number):
            number[i]=n/tot     # store the normalized fraction
        normed_name = ''

        for i in range(len(element)):
            normed_name+=element[i]
            digit = np.format_float_positional(round(number[i],3), trim='-') #include only 3 significant digit
            normed_name+=digit
            sum_comp+=float(digit)
    return element,sum_comp,normed_name  # output element, sum of composition to check if normalized
```

```python
#check formula 
if steel['formula'].isnull().values.any():
    print("ERROR: formula is missing")
else:
    print("No formula is missing")
    
print(len(np.where(steel['formula'].duplicated()==True)[0]),'duplicated data exist',)

    

#check yield strength
if steel['yield strength'].isnull().values.any():
    print('ERROR: yield strength of {} steels is missing'.
          format(len(np.where(steel['yield strength'].isnull().values==True)[0])),) 
else:
    print('No yield strength is missing')
print(steel['yield strength'].describe(),'\n') 
    

# check tensile strength
if steel['tensile strength'].isnull().values.any():
    print('ERROR: yield strength of {} steels is missing'.
          format(len(np.where(steel['yield strength'].isnull().values==True)[0])),) 
else:
    print('No tensile strength is missing')
print(steel['tensile strength'].describe(),'\n')  #the print result can indicate whether the series is pure numbers(floats)
    
# check elongation 
if steel['elongation'].isnull().values.any():
    print('ERROR: elongation data of {} steels is missing'.
          format(len(np.where(steel['elongation'].isnull().values==True)[0])),)    
else:
    print('No elongation is missing')
    
# filled the missing value with mean
steel = steel.fillna(steel['elongation'].mean())
print(steel['elongation'].describe(),'\n') 


# go over formula, collect elements, check if wt% sum up to 100 
norm_checker = 0
for i in steel['formula']:
    element,sum_comp,normed_name = comp_normalization(i)
    steel = steel.replace(i,normed_name)
#    steel['formula'][i] = normed_name
    if abs(sum_comp-1)<1e-3:
        norm_checker+=1
if norm_checker==len(steel['formula']):
    print('all compositions are normalized')
    
```

```python
steel2 = StrToComposition().featurize_dataframe(steel, "formula")
steel2
```

```python
#plot the correlation matrix
corr= steel2.corr().abs()\
                        .mask(np.tril(np.ones(corr.shape)).astype(bool)) # remove the lower triangle of values in the matrox 
corr.style.background_gradient(cmap='viridis')
```

```python
col_to_del = [feature for feature in corr if sum(corr[feature]>0.8)>0]
print(col_to_del)
steel3 = steel2.drop(col_to_del, axis=1)
steel3
```
